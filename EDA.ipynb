### **Perform EDA Operations :**

Step 1: Import Python Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

Step 2: Read Dataset

```python
df = pd.read_csv('used_cars_data.csv') # Read Data from the Dataset used_cars_data
df.head() # Return first 5 rows
```

```python
df.tail(15) # Return last 15 rows
```

```python
df.shape # Return the shape of the dataset in the form of tuple(no. of rows, no. of columns)
```

```python
df.columns # List all the columns present in the DataFrame
```

```python
df.dtypes # Return datatype of all the columns
```

Step 3: Summary Statistics

```python
df.describe() # Display Summary Statistics
```

```python
df.info()  # Display data types, null values, memory usage
```

```python
df.nunique()  #  Returns the number of unique values in each column of the DataFrame
```

Step 4: Dropping irrelevant columns

```python
df = df.drop(['Fuel_Type'] , axis =1)  # Drop irrelevant column
df.head()
```

Step 5: Renaming the columns

```python
df = df.rename(columns = {'Name': 'Model'}) # Rename the column 'Name' from 'Model'
df.head()
```

Step 6: Dropping the duplicate rows

```python
df.shape
```

```python
duplicate_rows_df = df[df.duplicated()]
print("number of duplicate rows: ", duplicate_rows_df.shape)  # Return the number of duplicated rows in the DataFrame
```

Step 7: Dropping the missing or null values

```python
df.isnull().sum()  # Count no. of null values in each column
```

```python
df = df.dropna()  # Drop null values
df.count()  # Count no. of rows in each columns
```

```python
print(df.isnull().sum())  # Print null values in the each column
```


Step 8: Detecting Outliers

```python
sns.boxplot(x=df['Price'])  # Boxplot for Price column
```

Step 9: Plot different features against one another

Histplot :

```python
plt.figure(figsize = (10,6))  # Plot Histplot
sns.histplot(df['Year'])
```

Scatter Plot :

```python
plt.figure(figsize = (10,6))
sns.scatterplot(x = df['Year'], y = df['Price']) # Plot Scatter plot between 'Year' and 'Price'
```

Line Plot :

```python
plt.figure(figsize = (10,6))
sns.lineplot(x = df['Year'], y = df['Price'])   # Plot Line plot between 'Year' and 'Price'
```

Heat Map :

```python
plt.figure(figsize = (10,6))
c = df.corr(numeric_only= True)
sns.heatmap(c, annot = True, cmap = 'Blues')  # Plot Heat Map
```
